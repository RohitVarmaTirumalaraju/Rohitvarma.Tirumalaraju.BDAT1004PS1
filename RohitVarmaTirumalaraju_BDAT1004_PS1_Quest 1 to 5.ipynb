```
//Question 3: Code executed in Visual Studio
//Answer

using System;
class TriangleArea
{
    static void Main(string[] args)
    {
        Console.WriteLine("Enter side a value:");
        int a = Convert.ToInt32(Console.ReadLine());
        Console.WriteLine("Enter side b value:");
        int b = Convert.ToInt32(Console.ReadLine());
        Console.WriteLine("Enter side c value:");
        int c = Convert.ToInt32(Console.ReadLine()); 
        double s = (a + b + c) / 2;
        double TAreasub = s * (s - a) * (s - b) * (s - c);
        double TArea = Math.Sqrt(TAreasub);
        Console.WriteLine($">>> TriangleArea({a},{b},{c})\n" + TArea);

    }

}
```

```
//Question 4: Code executed in Visual Studio
//Answer:

using System;

class EvenOdd
{
    static void Main(string[] args)
    {
        Console.Write("Input the number of elements to be stored in the array :");
        int num = Convert.ToInt32(Console.ReadLine());
        int evennum = 0;
        int oddnum = 0;
        Console.WriteLine($"Input {num} elements in the array : ");
        int[] elements = new int[num];
        int[] even = new int[num];
        int[] odd = new int[num];
        int izero = 0;
        while (izero < num)
        {
            Console.Write($"element - {izero} : ");
            int k = Convert.ToInt32(Console.ReadLine());
            elements[izero] = k;
            izero++;
        }
        foreach (int number in elements)
        {
            if (number % 2 == 0)
            {
                even[evennum] = number;
                evennum = evennum + 1;
            }
            else
            {
                odd[oddnum] = number;
                oddnum = oddnum + 1;
            }
        }
        Console.Write("The Even elements are: \n");
        int x=0;
        while (x < evennum)
        {
            Console.Write(even[x] + " ");
            x++;
        }
        Console.Write("\nThe Odd elements are: \n");
        int y=0;
        while (y < oddnum)
        {
            Console.Write(odd[y] + " ");
             y++;
        }
    }
}


```
